```python
### axes method[left,bottom,hight,width]fraction
ax=plt.axes([0.1,0.1,0.8,0.8])
x = np.arange(1,10,0.2)
s = np.sin(x)
c = np.cos(x)
ax.plot(x,s,'g+:')
ax.plot(x,c,'rH-')
ax.legend(['RCB','CSK'],loc="upper right")
ax.set_title("IPL Match")

```

```python
x = [1,2,3,4]
y = [3,1,7,1]
fg,ax = plt.subplots(2,2)
ax[0][0].plot(x,y)
ax[0][1].scatter(x,y)
ax[1][0].bar(x,y)
ax[1][1].hist(x)
ax[0][0].set_title("My line graph")
ax[0][1].set_title("My scatter graph")
ax[1][0].set_title("My bar graph")
ax[1][1].set_title("My histogram graph")
plt.show()
```

```python
x= np.array([1,2,3,4])
fg,ax=plt.subplots(2,2)
ax[0][0].plot(x)
ax[0][1].plot(x**2)
ax[1][0].plot(x**3)
ax[1][1].plot(x**(0.5)) ### square root
```

```python
#### fill color function(fill_between)
fg,ax=plt.subplots(2,1,figsize=(5,5))
x = np.arange(0,10,0.1)
y = np.sin(x)
ax[0].plot(x,y)
ax[0].fill_between(x,y,0)
ax[1].plot(x,y)
ax[1].fill_between(x,y,-1)
```

```python
#### bar graph 


import random 
m = [random.randint(0,101)for i in range(5)]
sn=('Ram','shyam','akash','abhinav','arsh')
plt.bar(sn,m,width=0.2,color='r')
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("student marks")
plt.show()
```

```python
### horizontally bar graph

import random 
m = [random.randint(0,101)for i in range(5)]
sn=('Ram','shyam','akash','abhinav','arsh')
plt.barh(sn,m,height=0.2,color='r')
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("student marks")
plt.show()
```

```python
### TEXT FUNCTION

import random 
m = [random.randint(0,101)for i in range(5)]
sn=('Ram','shyam','akash','abhinav','arsh')
plt.bar(sn,m,width=0.2,color='r')
for i in range(len(sn)):
    plt.text(i,m[i]+1.5,str(m[i]),color="b")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("student marks")
plt.show()
```

```python
### horizontAL TEXT FUNCTION
import random 
m = [random.randint(0,101)for i in range(5)]
sn=('Ram','shyam','akash','abhinav','arsh')
plt.barh(sn,m,height=0.2,color='r')
for i in range(len(sn)):
    plt.text(m[i]+1.5,i,str(m[i]),color="b")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("student marks")
plt.show()
```

```python
import random 
math = [random.randint(0,101)for i in range(5)]
eng = [random.randint(0,101)for i in range(5)]
hindi = [random.randint(0,101)for i in range(5)]
names=['a','b','c','d','e']
width=0.3
plt.figure(figsize=(5,5))

MathIdx = [i for i in range(5)]
EngIdx=[i + width for i in range(5)]
HindiIdx=[i + 2*width for i in range(5)]

plt.bar(MathIdx,math,color='r',width=width)
plt.bar(EngIdx,eng,color='b',width=width)
plt.bar(HindiIdx,hindi,color='k',width=width)
plt.legend(['Math','English','Hindi'],loc='upper right')
plt.show()
```

### PIE CHART

```python
import random
x=[random.randint(30,101) for i in range(5)]
y =['math','eng','chem','phy','hindi']
plt.pie(x,labels=y,colors=['g','r','k','y','b'],shadow='gray')
plt.show
```

```python
### explode function
import random
x=[random.randint(30,101) for i in range(5)]
y =['math','eng','chem','phy','hindi']
ex=[0.2,0,0,0,0]
plt.pie(x,labels=y,colors=['g','r','k','y','b'],shadow='gray',explode=ex)
plt.show
```

```python
### starting angle
import random
x=[random.randint(30,101) for i in range(5)]
y =['math','eng','chem','phy','hindi']
ex=[0.2,0,0,0,0]
plt.pie(x,labels=y,colors=['g','r','k','y','b'],shadow='gray',explode=ex,startangle=90)
plt.show
```
