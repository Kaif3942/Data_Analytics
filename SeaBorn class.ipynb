```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=sns.load_dataset("tips")
print(df)
```

```python
sns.stripplot(x='tip',y='sex',data=df)
plt.plot()
```

```python
x = [1,2,3,4,5]
y =[9,0,4,7,2]
ds = pd.DataFrame({'a':x,'b':y})
sns.stripplot(ds)
```

```python
plt.figure(figsize=(5,2))
sns.scatterplot(x='tip', y='total_bill',data=df,hue='day')
```

```python
plt.figure(figsize=(6,3))
color =['green','blue','Red','Yellow']
sns.set_palette(color)
sns.lineplot(x='tip',y='total_bill',data=df,marker='D',hue='day')
```

```python
plt.figure(figsize=(6,3))
color =['green','blue','Red','Yellow']
sns.set_palette(color)
sns.set_style("white")
sns.lineplot(x='tip',y='total_bill',data=df,marker='D',hue='day')
```

```python
State=['Rajastha','Bihar','Madhya pradesh',
       'gujrat','maharastra']
growth =[342239,94163,308245,196024,307713]

df=pd.DataFrame({"State":State,
                "Growth":growth})
df.sort_values('Growth',axis=0,ascending=True,inplace=True)
print(df)
sns.barplot(x='State',y='Growth',data=df)
```

### swarm plot

### Histogram plot
